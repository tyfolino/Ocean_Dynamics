```python
import xgcm
metrics = {
    ('X',): ['dxC', 'dxG'], # X distances
    ('Y',): ['dyC', 'dyG'], # Y distances
    ('Z',): ['drF', 'drC'], # Z distances
    ('X', 'Y'): ['rA', 'rAz', 'rAs', 'rAw'] # Areas
}
grid = xgcm.Grid(ds, metrics=metrics,periodic=False)
grid
```

```python
U_center = grid.interp(ds.U, 'X', boundary='fill')
V_center = grid.interp(ds.V, 'Y', boundary='fill')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ds.Eta[-1].plot(ax=ax)
ax.quiver(ds.XC, ds.YC, U_center[-1], V_center[-1])
```

$$ \psi(x,y,t) = -\int_0^y dy' \int_{-H}^0 dz u(x', y, t) $$ 

The following line is zonal current * layer thickness (drF = 5000 m)

```python
ψ = -grid.cumint((ds.U * ds.drF[0]),'Y',boundary='fill')/1e6
ψ
```

```python
ψ.mean(dim='time').plot()
```

```python
Γ=grid.diff(ds.V * ds.dyC, 'X', boundary='extend') - grid.diff(ds.U * ds.dxC, 'Y', boundary='extend')
ζ = Γ / ds.rAz
ζ[-1].plot()
```

```python
τ0 = 0.1/1000
pi = np.pi
β = 1e-11 # s^-1 m^-1
x = ds.XG
a = float(ds.dxC.isel(YC=0,XG=slice(1,-1)).sum(dim='XG'))
r = 1e-6
ϵ = r / (β * a)
y = ds.YG
```

```python
ψ_sto = (τ0 * pi / β) * (1 - (x/a) - np.exp((-1*x) / (a * ϵ))) * np.sin(pi * y / a) / 1e6
```
